## A1. 답안 작성  
답변: A/B 테스트는 두 가지 버전(A안, B안)을 사용자에게 무작위로 노출시킨 후, 클릭률이나 전환율과 가튼 성과 지표를 비교하여 어떤 안이 더 효과적인지를 통계적으로 검증하는 실험 기법이다.

A/B 테스트를 수행할 때에는 실험 결과가 왜곡되지 않도록 여러 요인을 고려해야한다.

첫째, 무작위 할당(Random Assignment)이 반드시 이루어져야 한다. 각 A/B안에 사용자를 임의로 배정하지 않으면, 사용자 특성 차이로 인해 결과가 편향될 수 있다.

둘째, 한 번의 실험에서는 한 가지 요소만 변경해야 한다. 여러 요소를 동시에 변경할 경우, 성과 차이가 어떤 요인에 의해 발생했는지 명확히 해석하기 어렵다.

셋째, 초두효과(Primacy Effect)를 고려하여 충분한 표본 수와 실험 기간을 확보해야 한다. 초두효과란 사용자가 처음 접한 정보나 디자인에 더 큰 영향을 받는 현상으로, 새로운 UI나 기능이 처음에는 호기심으로 인해 클릭률이나 이용률이 일시적으로 높게 나타날 수 있다. 이럴 경우 단기 성과만으로 판단하면 과대평가의 위험이 있다. 따라서 충분한 표본 수를 확보하고 일정 기간 이상 실험을 유지하여 효과의 지속성을 확인해야 한다.

넷째, 성과지표(KPI)를 명확히 정의해야 한다. 단순 클릭률 상승이 실제 전환이나 서비스 목표 달성으로 이어지는지 함께 고려하지 않으면 의미 없는 개선으로 오해할 수 있다.

마지막으로, 실험 결과는 반드시 통계적 유의성을 검증 후 해석해야 한다. 관측된 차이가 우연에 의한 것인지, 실제 효과인지를 통계적으로 판단하지 않으면 잘못된 의사결정을 내릴 가능성이 있다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

*자동으로 계산해주는 코드 쓰지 말기-오답처리할 것임

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:
p1은 A(Control)에서의 성공 확률을 의미한다. p2는 B(Variant)에서의 성공 확률을 의미한다. 본 실험에서는 공모전 상세 페이지 진입률(Detail Page CVR)로 p1,p2를 가정하였다.

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.
import math
from scipy.stats import norm

# 1. 유의수준과 검정력 설정
alpha = 0.05
power = 0.8
beta = 1-power

# 2. p1,p2 설정(Detail Page CVR)
p1 = 0.1
p2 = 0.12

# 3. Z값 계산(양측 검정)
z_alpha_over_2 = norm.ppf(1-alpha /2)
z_beta = norm.ppf(1-beta)

# 4. 표본 수 계산(문제에 제시된 공식 사용)
numerator = (z_alpha_over_2 + z_beta)**2 * (p1*(1-p1)+ p2*(1-p2))
denominator = (p1-p2)**2
n= numerator/denominator

# 5. 각 그룹당 최소 표본 수
n_per_group = math.ceil(n)


print("각 그룹당 최소 test users 수:", n_per_group)
print("총 test users 수:", 2 * n_per_group)


#########
```

[실험 가설]  
귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$

## Q3-1. 위 가설에 적합한 검정 이름과 왜 해당 검정을 써야하는지 이유도 작성해주세요. (t test와 카이제곱검정 중)

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 **카이제곱 검정(Chi-square test)** 이고,  

이유는 비교하고자 하는 지표(CVR)가 연속형이 아닌 '진입/미진입'로 정의되는 비율이기 때문이다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

실제 데이터가 없어서, 저도 필요한 데이터는 가정했답니다.

## A3-2. 아래 두 가설 내용을 채워주세요.
  
[실험 가설]  
귀무가설 ($H_0$): 대조군(A)과 실험군(B)의 공모전 상세 페이지 진입률은 동일하다. 즉, 색상 표현 여부와 상세 페이지 진입률은 무관하다.

$$
H_0 : p_A = p_B
$$


  
대립가설 ($H_1$): 대조군(A)과 실험군(B)의 공모전 상세 페이지 진입률은 유의한 차이를 보인다. 즉, 색상 표현이 상세 페이지 진입률에 영향을 미친다.(증가 또는 감소)
$$
H_1 : p_A \neq p_B
$$

*마감 여유도 색상 표현은 사용자에게 긍정적 또는 부정적 신호로 작용할 수 있으므로, 본 실험에서는 효과의 방향을 사전에 가정하지 않고 상세 페이지 진입률의 변화 여부를 검정하기 위해 양측 가설을 설정하였다.

그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성
import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력(가정값)
control_results = [384, 3455]  # 대조군(A)
variant_results = [430, 3409]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행(양측 검정)
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 마감 여유도 색상 표현은 공모전 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 마감 여유도 색상 표현의 효과가 통계적으로 유의하다고 보기 어렵습니")










#############
```

**결과해석**
